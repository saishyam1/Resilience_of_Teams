```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random   
import scipy
from scipy import stats
from scipy.stats import kurtosis
from scipy.stats import skew
```

```python
def BiC(dataframe,node):
    n=len(dataframe)
    sk=skew(dataframe.loc[:,node],axis=0,bias=True)
    kur=kurtosis(dataframe.loc[:,node],bias=True,axis=0,fisher=True)
    ncum=(pow(n-1,2))/((n-2)*(n-3))
    bic=(pow(sk,2) + 1)/(kur +3*ncum)
    return bic
```

```python
def Boolean_file_import(topo_file_name):
    csv_file_name = topo_file_name[:-5] + "_finFlagFreq.csv"
    nodes_file_name = topo_file_name[:-5]  + "_nodes.txt"
    data = pd.read_csv(csv_file_name)

    new_columns = data['states'].str.split('_', expand=True)
    new_columns.columns = np.array(pd.read_csv(nodes_file_name, header=None).T)[0]
    
    df = pd.concat([data, new_columns], axis=1)
    return df
```

```python

def average_finder(dataframe, Team1_Nodes, Team2_Nodes, Average_Team1, Average_Team2):
    dataframe.loc[:,Team1_Nodes + Team2_Nodes] = dataframe.loc[:,Team1_Nodes + Team2_Nodes].astype(int)
    dataframe.loc[:,Average_Team1] = dataframe.loc[:,Team1_Nodes].mean(axis = 1)
    dataframe.loc[:,Average_Team2] = dataframe.loc[:,Team2_Nodes].mean(axis = 1)
    dataframe.loc[:,'frequency'] = dataframe.loc[:,"Avg0"]*100000
    dataframe.loc[:,'frequency'] =dataframe.loc[:,'frequency'].astype(int)
    dataframe = dataframe.sort_values(by='Avg0', ascending=False)

    return dataframe

```

```python
def f1_score_new_any_dimension_threshold_toggle_switch_optimized_boolean(df, Team_0_state,Team_1_state ):
    single_pos = df[((df[Team_0_state] == 1) & (df[Team_1_state] == 0)) | ((df[Team_0_state] == 0) & (df[Team_1_state] == 1))]
    return single_pos.loc[:,'Avg0'].sum()
```

```python
import pandas as pd

def duplicate_rows1(df):
    new_index = df.index.repeat(df['frequency'])
    duplicated_df = df.loc[new_index].reset_index(drop=True)
    duplicated_df = duplicated_df.drop(columns=['frequency'])
    return duplicated_df
```

```python
def percentage_impurity_finder(topo,team1,team2):
    impurity = 0
    matrix =  matrix_index_sort(Topo_adj(topo))
    teams = [team1,team2]
    for i in range(len(teams)):
        for j in range(len(teams)):
            
            sub_matrix = matrix.loc[teams[i],teams[j]]
            
            if i == j:
                 impurity += (sub_matrix.loc[teams[i],teams[j]] == -1).sum().sum()
            else:
                impurity += (sub_matrix.loc[teams[i],teams[j]] == 1).sum().sum()


    return impurity/(len(team1+team2)**2)
```

```python
def embedded_team_identifies(topo,perisize):
    matrix = matrix_index_sort(Topo_adj(topo))
    total_nodes = len(matrix)
    core_size = total_nodes - perisize

    core_to_corner_matrix = centreing_index_list(list(matrix.columns), core_size)
    Team1 = core_to_corner_matrix[perisize:perisize+int(core_size/2)]
    Team2 = core_to_corner_matrix[perisize+int(core_size/2):perisize+core_size]
    return Team1, Team2
```

```python
def Team_Strength_finder(influence_matrix, Team1, Team2):   
    TeamScores = pd.DataFrame(0, columns = [0,1], index = [0,1] )

    teams = [Team1, Team2]

    for t1 in range(len(teams)):
        for t2 in range(len(teams)):
            Tsc= 0 
            count = 0
            #print(teams[t1])
            for i in teams[t1]:
                for j in teams[t2]:
                    Tsc += influence_matrix.loc[i,j]
                    count +=1
                    #print(t1,t2)
            TeamScores.loc[t1,t2] = Tsc/count   

    TeamScores.columns = ['TeamB','TeamA']

    TeamScores.index = ['TeamB','TeamA']
    TeamScores
    return abs(TeamScores).sum(axis = 1).sum(axis = 0)/4
```

```python
def influence(dp, n,pl):
    n=len(set(dp['Source']))
    infl = pd.DataFrame([[0]*n]*n)
    infl.columns=set(dp.loc[:,'Source'])
    infl.index=set(dp.loc[:,'Source'])
    adj = Topo_adj(dp)
    adj=matrix_index_sort(adj)

    maxi = abs(adj)
    pr = adj
    mr = abs(adj)
    infl = adj
    for i in range(1, pl):
        pr=adj@pr
        mr = maxi@mr
        qwe = pr/mr

        infl = infl + qwe.replace(np.nan, 0)
        
        

    infl  = infl/pl
    infl=infl.round(2)
    return infl
```

```python
def Team_Strength_finder(influence_matrix, Team1, Team2):   
    TeamScores = pd.DataFrame(0, columns = [0,1], index = [0,1] )

    teams = [Team1, Team2]

    for t1 in range(len(teams)):
        for t2 in range(len(teams)):
            Tsc= 0 
            count = 0
            #print(teams[t1])
            for i in teams[t1]:
                for j in teams[t2]:
                    Tsc += influence_matrix.loc[i,j]
                    count +=1
                    #print(t1,t2)
            TeamScores.loc[t1,t2] = Tsc/count   

    TeamScores.columns = ['TeamB','TeamA']

    TeamScores.index = ['TeamB','TeamA']
    TeamScores
    return abs(TeamScores).sum(axis = 1).sum(axis = 0)/4
```

```python
def BiC_CC_F1_team_strength_Boolean(folder, TS1 , TS2 , filee , n_files ,
                            TS1_nodes, TS2_nodes , Self_act_status ,start = 0, end=100):
    # print(folder)
    core_size = len(TS1_nodes+ TS2_nodes)

    global melt_table_complete

    Corr_list = []
    F1_list = []
    density_list =[]
    impurity_list =[]
    TS_list_core = []
    indices = []
    team_size = []
    IndegreeA = []
    IndegreeB = []
    OutdegreeA = []
    OutdegreeB = []
    Mean_Connectivity = []
    counts_of_self_act = []
    cols = ['index','Core Density','CC AB','F1','Team_Strength',
                                 'Team Size']


    temp_table_complete=pd.DataFrame([[0]*len(cols)]*n_files)
    temp_table_complete.columns = cols
    

    temp_table_complete.loc[:,'Self Activation Status']=Self_act_status



    for i in range(start, end):
        # print(i)


        #F1 Calculation
        topo_file_name =  folder+ '/'+filee+  str(i) +'.topo'
        dataframe_ = Boolean_file_import(topo_file_name)
        dataframe_ = average_finder(dataframe_, TS1_nodes, TS2_nodes, TS1,TS2)
        F1_list.append( f1_score_new_any_dimension_threshold_toggle_switch_optimized_boolean(
            dataframe_, TS1,TS2 
            ))

        #CC AB Calculation
        dataframe_ = duplicate_rows1(dataframe_)
        Corr_list.append(stats.spearmanr(dataframe_.loc[:,TS1], dataframe_.loc[:,TS2])[0])

        #Index Addition
        indices.append(filee+ '_' + str(i))
        
        #Team Strength Calculation
        influ = influence(pd.read_csv(topo_file_name, sep = '\t'), 1,10)
        Core_A, Core_B =  TS1_nodes, TS2_nodes
        TS_list_core.append(Team_Strength_finder(influ, Core_A, Core_B))
        team_size.append(len(TS1_nodes))
        print(topo_file_name)
        


        topo_file = pd.read_csv(topo_file_name, sep = '\t')
        # return topo_file
        net_density = core_density_finder(matrix_index_sort(Topo_adj(topo_file)), Core_A+Core_B )
        density_list.append(net_density*100)
        impurity_list.append(percentage_impurity_finder(topo_file,Core_A, Core_B)*100/net_density)

        adjmat_core_peri=centreing_core(matrix_index_sort(Topo_adj(topo_file)),core_size)
        total_nodes = len(adjmat_core_peri)
        peri_nodes = total_nodes - core_size

        IndegreeA.append( len(topo_file[(topo_file['Target'].isin(Core_A)) & ~(topo_file['Source'].isin( Core_A + Core_B ))]))
        # print(len(topo_file[(topo_file['Target'].isin(Core_A)) & ~(topo_file['Source'].isin( Core_A + Core_B ))]),Core_A, Core_A + Core_B)
        IndegreeB.append(len(topo_file[(topo_file['Target'].isin(Core_B)) & ~(topo_file['Source'].isin( Core_A + Core_B ))]))
        OutdegreeA.append(len(topo_file[(topo_file['Source'].isin(Core_A)) & ~(topo_file['Target'].isin( Core_A + Core_B ))]))
        OutdegreeB.append(len(topo_file[(topo_file['Source'].isin(Core_B)) & ~(topo_file['Target'].isin( Core_A + Core_B ))]))
        Mean_Connectivity.append(adjmat_core_peri.iloc[:peri_nodes,:peri_nodes].abs().sum().sum()/peri_nodes)
    

        

    # print(Corr_list)
    temp_table_complete.loc[:,'CC AB']= Corr_list
    temp_table_complete.loc[:,'F1']=F1_list
    temp_table_complete.loc[:,'Impurity']=impurity_list
    temp_table_complete.loc[:,'Core Density']=density_list

    temp_table_complete.loc[:,'Team_Strength']=TS_list_core
    temp_table_complete.loc[:,'index']=indices
    temp_table_complete.loc[:,'Team Size']= team_size


    temp_table_complete.loc[:,'Indegree_Team_A'] = IndegreeA
    temp_table_complete.loc[:,'Indegree_Team_B'] = IndegreeB
    temp_table_complete.loc[:,'Outdegree_Team_A'] = OutdegreeA
    temp_table_complete.loc[:,'Outdegree_Team_B'] = OutdegreeB
    temp_table_complete.loc[:,'External Network Density'] = Mean_Connectivity






    # temp_table_complete.loc[:,'Number_of_Self_Activations']= counts_of_self_act



    melt_table_complete=pd.concat([melt_table_complete, temp_table_complete])


    
    return  melt_table_complete



```

```python
def BiC_CC_F1_team_strength_Boolean( folder, TS1 , TS2 , filee , n_files , TS1_nodes, TS2_nodes , Self_act_status ,rp, 
                                    start = 0, end=100):
    
    print(folder)
    core_size = len(TS1_nodes+ TS2_nodes)

    global melt_table_complete

    Corr_list = []
    F1_list = []
    density_list =[]
    impurity_list =[]
    TS_list_core = []
    indices = []
    team_size = []
    IndegreeA = []
    IndegreeB = []
    OutdegreeA = []
    OutdegreeB = []
    Mean_Connectivity = []
    counts_of_self_act = []
    cols = ['index','Core Density','CC AB','F1','Team_Strength',
                                 'Team Size']


    temp_table_complete=pd.DataFrame([[0]*len(cols)]*n_files)
    temp_table_complete.columns = cols
    

    temp_table_complete.loc[:,'Self Activation Status']=Self_act_status



    for i in range(start, end):
        print(i)


        #F1 Calculation
        topo_file_name =  folder+ '/'+filee+  str(rp) +'.topo'
        dataframe_ = Boolean_file_import(topo_file_name)
        dataframe_ = average_finder(dataframe_, TS1_nodes, TS2_nodes, TS1,TS2)
        F1_list.append( f1_score_new_any_dimension_threshold_toggle_switch_optimized_boolean(
            dataframe_, TS1,TS2 
            ))

        #CC AB Calculation
        dataframe_ = duplicate_rows1(dataframe_)
        Corr_list.append(stats.spearmanr(dataframe_.loc[:,TS1], dataframe_.loc[:,TS2])[0])

        #Index Addition
        indices.append(filee+ '_' + str(rp))
        
        #Team Strength Calculation
        influ = influence(pd.read_csv(topo_file_name, sep = '\t'), 1,10)
        Core_A, Core_B =  TS1_nodes, TS2_nodes
        TS_list_core.append(Team_Strength_finder(influ, Core_A, Core_B))
        team_size.append(len(TS1_nodes))
        


        topo_file = pd.read_csv(topo_file_name, sep = '\t')
        net_density = core_density_finder(matrix_index_sort(Topo_adj(topo_file)), Core_A+Core_B )
        density_list.append(net_density*100)
        impurity_list.append(percentage_impurity_finder(topo_file,Core_A, Core_B)*100/net_density)

        adjmat_core_peri=centreing_core(matrix_index_sort(Topo_adj(topo_file)),core_size)
        total_nodes = len(adjmat_core_peri)
        peri_nodes = total_nodes - core_size

        IndegreeA.append( len(topo_file[(topo_file['Target'].isin(Core_A)) & ~(topo_file['Source'].isin( Core_A + Core_B ))]))
        IndegreeB.append(len(topo_file[(topo_file['Target'].isin(Core_B)) & ~(topo_file['Source'].isin( Core_A + Core_B ))]))
        OutdegreeA.append(len(topo_file[(topo_file['Source'].isin(Core_A)) & ~(topo_file['Target'].isin( Core_A + Core_B ))]))
        OutdegreeB.append(len(topo_file[(topo_file['Source'].isin(Core_B)) & ~(topo_file['Target'].isin( Core_A + Core_B ))]))
        Mean_Connectivity.append(adjmat_core_peri.iloc[:peri_nodes,:peri_nodes].abs().sum().sum()/peri_nodes)
    

        

    print(Corr_list)
    temp_table_complete.loc[:,'CC AB']= Corr_list
    temp_table_complete.loc[:,'F1']=F1_list
    temp_table_complete.loc[:,'Impurity']=impurity_list
    temp_table_complete.loc[:,'Core Density']=density_list

    temp_table_complete.loc[:,'Team_Strength']=TS_list_core
    temp_table_complete.loc[:,'index']=indices
    temp_table_complete.loc[:,'Team Size']= team_size


    temp_table_complete.loc[:,'Indegree_Team_A'] = IndegreeA
    temp_table_complete.loc[:,'Indegree_Team_B'] = IndegreeB
    temp_table_complete.loc[:,'Outdegree_Team_A'] = OutdegreeA
    temp_table_complete.loc[:,'Outdegree_Team_B'] = OutdegreeB
    temp_table_complete.loc[:,'External Network Density'] = Mean_Connectivity






    # temp_table_complete.loc[:,'Number_of_Self_Activations']= counts_of_self_act



    # melt_table_complete=pd.concat([melt_table_complete, temp_table_complete])


    
    return  temp_table_complete



```

```python
import multiprocessing


def process_parameter_set(parameters):
    return BiC_CC_F1_team_strength_Boolean(*parameters)

if __name__ == "__main__":
    # Define the number of cores
    num_cores = 15

    # Create a multiprocessing pool with the number of desired cores
    pool = multiprocessing.Pool(processes=num_cores)

    # Define the range of rp values to distribute among cores (0 to 99)
    rp_range = range(2500)

    # Create a list of input parameter sets for each unique rp

    parameter_sets = [    ( 'All_vary/all_vary', 'Average_15_16_17_18_19', 
                           'Average_20_21_22_23_24', 
                            'Random_Network_TS_5_embed_P30N_C10N_trial',1,
                            ["N15","N16","N17","N18","N19"], ["N20","N21","N22","N23","N24"] ,  '1',rp, 0, 1   )  for rp in rp_range  ] 

 

    # Use the pool.map() function to parallelize the processing for each unique rp
    results = pool.map(process_parameter_set, parameter_sets)

    # Close the pool to release resources
    pool.close()
    pool.join()


    melt_table_complete_run = pd.concat(results, ignore_index=True)
    # melt_table_complete=pd.concat([melt_table_complete,melt_table_complete_run])
```
